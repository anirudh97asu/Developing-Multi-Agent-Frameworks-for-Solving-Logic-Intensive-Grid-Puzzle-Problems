```python
import re
def process_ls(ls):
    if re.search(r"\d+",ls[0]) is None:
        return ls

    else:
        res = []
        for val in ls:
            val = re.sub(r"[^\d+]", '', val)
            res.append(val)
        return res
```

```python
import re
def process_question(question):
    lines = question.split("\n")
    categories = {}
    clues = []
    actual_question = ""
    ckpt = -1
    flag = -1
    for index, line in enumerate(lines):


        if ckpt == 1 and re.search(r"^\d+\.", line) is None:
            break

        if " : " in line:
            if flag == -1:
                flag = index
            title, ls = line.split(" : ")[0], line.split(" : ")[1].split(", ")
            ls = process_ls(ls)
            categories[title] = ls
        
        if "Clues" in line:
            ckpt = 1
            continue
        
        if ckpt == 1 and re.search(r"\d+\.", line) is not None:
            clues.append(line)
        


    actual_question += "\n".join(lines[:flag])            

    return actual_question, categories, clues


ques, categories, clues = process_question(df["question"].iloc[100])
```

```python
from constraint_tools import ConstraintSolver
cs = ConstraintSolver(categories=categories)
 
methods = [(method, getattr(ConstraintSolver, method)) for method in dir(ConstraintSolver) if callable(getattr(ConstraintSolver, method))and not method.startswith('__')]

tool_descriptions = {}
for method in methods:
    
    if method[0] not in ["create_category_vars", "set_category_domain", "add_distinct"]:
        tool_descriptions[method[0]] = method[1].__doc__
    
    else:
    
        cs.create_category_vars
        for category in categories:
            
            cs.create_category_vars(category)
    
            min_val, max_val = 1, len(categories[category])
            cs.set_category_domain(category, lower=min_val, upper=max_val)
            
            cs.add_distinct(category)
    

print(len(tool_descriptions))
```

```python
# Create grid with Python - simple brute force approach
# This should work for most logic grid puzzles

months = ["January", "February", "March", "April", "May"]
fliers = ["Brandi", "Christie", "Eloise", "Florence", "Jaime"]
charms = ["coin", "lucky_hat", "shamrock", "talisman", "wishbone"]

from itertools import permutations

# Try all possible arrangements of people and charms
for month_to_flier in permutations(fliers, len(months)):
    # Create month -> flier mapping
    month_flier_map = {months[i]: month_to_flier[i] for i in range(len(months))}
    
    for month_to_charm in permutations(charms, len(months)):
        # Create month -> charm mapping
        month_charm_map = {months[i]: month_to_charm[i] for i in range(len(months))}
        
        # Check all constraints
        valid = True
        
        # 1. The passenger leaving in February will bring their lucky hat.
        if month_charm_map["February"] != "lucky_hat":
            valid = False
            continue
        
        # 2. Christie will leave 1 month after Eloise.
        eloise_month_index = months.index(next(m for m, f in month_flier_map.items() if f == "Eloise"))
        christie_month_index = months.index(next(m for m, f in month_flier_map.items() if f == "Christie"))
        if christie_month_index != eloise_month_index + 1:
            valid = False
            continue
        
        # 3. The passenger with the wishbone will leave 2 months after Eloise.
        wishbone_month_index = months.index(next(m for m, c in month_charm_map.items() if c == "wishbone"))
        if wishbone_month_index != eloise_month_index + 2:
            valid = False
            continue
        
        # 4. The aerophobe leaving in May is either the aerophobe with the talisman or the aerophobe with the shamrock.
        may_charm = month_charm_map["May"]
        if may_charm != "talisman" and may_charm != "shamrock":
            valid = False
            continue
        
        # 5. The flier with the coin will leave sometime after the passenger with the shamrock.
        coin_month_index = months.index(next(m for m, c in month_charm_map.items() if c == "coin"))
        shamrock_month_index = months.index(next(m for m, c in month_charm_map.items() if c == "shamrock"))
        if coin_month_index <= shamrock_month_index:
            valid = False
            continue
        
        # 6. Of the flier with the coin and the passenger with the talisman, one is Brandi and the other will leave in May.
        coin_flier = month_flier_map[next(m for m, c in month_charm_map.items() if c == "coin")]
        talisman_flier = month_flier_map[next(m for m, c in month_charm_map.items() if c == "talisman")]
        talisman_month = next(m for m, c in month_charm_map.items() if c == "talisman")
        
        condition_1 = coin_flier == "Brandi" and talisman_month == "May"
        condition_2 = talisman_flier == "Brandi" and month_charm_map["May"] == "coin"
        
        if not (condition_1 or condition_2):
            valid = False
            continue
        
        # 7. Florence won't leave in March.
        if month_flier_map["March"] == "Florence":
            valid = False
            continue
        
        # If we get here, we've found a valid solution
        if valid:
            print("Solution found!")
            print("Month\t| Person\t| Lucky Charm")
            print("-" * 45)
            for month in months:
                flier = month_flier_map[month]
                charm = month_charm_map[month]
                print(f"{month}\t| {flier}\t\t| {charm}")
            
            # Only print the first valid solution
            break
    
    # Break out of outer loop if solution found
    if valid:
        break
else:
    print("No solution found!")
```

```python
import json
import traceback

max_iterations = 120
last_response = None
iteration = 0
iteration_response = []

system_prompt = f"""
You are an expert AI assistant for solving logic puzzles using the z-solver library. Your task is to systematically translate puzzles into z-solver functions following these precise steps:
Core Process:
    1. Carefully analyze all categories, items, and constraints in the puzzle
    2. Convert each clue into appropriate z-solver function calls
    3. Build the solution sequence methodically, one function at a time
    4. Respond with exactly one function call per message
Technical Requirements:
    - Always build constraints by operating on variables first (e.g., `equal(varA, varB)`)
    - Only use logical operators like `Or()` and `And()` to combine existing constraints
    - Never apply logical operators directly to raw variables
    - Maintain strict type adherence as required by each z-solver function
    - Verify constraint logic before providing each function call
Interaction Format:
    - Wait for user confirmation before proceeding to the next function
    - Provide only the function call without explanations unless requested
    - If uncertain about a clue's interpretation, ask for clarification

Begin when presented with a complete logic puzzle, and proceed systematically until the puzzle is solved.
The tool descriptions are given below:\n
"""

for key, value in tool_descriptions.items():
    key = key.strip()
    value = value.strip()
    system_prompt += f"""{key}: {value}"""

query = f"""{system_prompt}\n\n{ques}]\n\n{clues}\n\n{categories}"""

flag = -1

while iteration < max_iterations:
    iteration_result = None

    if flag == 1:
        print("success")
        break
    elif flag == 2:
        print("error")
        break

    print(f"\n--- Iteration {iteration + 1} ---")
    if last_response == None:
        current_query = query
    else:
        current_query = current_query + "\n\n" + " ".join(iteration_response)
        current_query = current_query + "  What should I do next?"

    # Get model's response
    response = agent.perform_action(query)

    
    print(f"Agent Response: {response}")

    call = json.loads(response)

    
    name = call["function_name"]
    params = call["parameters"]
    print(type(params))

    if re.search(r"and|or", name) is not None:
        print(name, params)
        continue

    if name == "model" or name == "check":
        flag = 1

    method = getattr(cs, name)
    annotations = method.__annotations__
    annot = {k:v for k,v in annotations.items() if k != "return"}

    if len(annot) == 0:
        iteration_result = method()
    
    elif len(annot) == len(params):
        try:
            iteration_result = method(*params)
        
        except Exception as e:
            print("error case_2", name, params, traceback.format_exc(e))
            flag = 2

    else:
        try:
            iteration_result = method(*params)
        
        except Exception as e:
            print("error case 3", name, params, traceback.format_exc(e))
            flag = 2

    print(f"  Result: {iteration_result}")
    cs.add_constraint_to_solver(iteration_result)
    last_response = iteration_result
    iteration_response.append(f"In the {iteration + 1} iteration you called {name} with {params} parameters, and the function returned {iteration_result}.")

    iteration += 1
    
```
